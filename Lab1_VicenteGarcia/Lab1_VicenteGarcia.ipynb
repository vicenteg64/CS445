# Lab 1 Lab Description
This dataset is titled "Wine Quality." It features a variety of wine from Portugal from a brand called "Vinho Verde" wine. The data is separated into two CSV files for the red wine and white wine. This data features attributes mostly based on physicochemical tests. This features attributes like total sulfur dioxide, residual sugar, density, and quality, which are measured continuously. This data set is said to feature more wines rated to have good quality than bad.

Here are some of the data from both data sets:

```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline

#reading the database
redwine_tips = pd.read_csv("winequality-red.csv",  sep=';') #add the separtor to read the data correctly
whitewine_tips = pd.read_csv("winequality-white.csv",  sep=';')
 
#printing the top 10 rows
display(redwine_tips.head(10))
display(whitewine_tips.head(10))
```

### Bar Graph Visualization

```python
import numpy as np #import numpy arrays

#Make a data frame that only has the top 5 red and white whines in quality
top5_red = redwine_tips.nlargest(5, 'quality')
top5_white = whitewine_tips.nlargest(5, 'quality')
wine_nums = [1, 2, 3, 4, 5]


#Assign the size of the plot figure
fig = plt.figure(figsize=(10, 6))
#Get only the values for the sulfer dioxide
plt.bar(wine_nums, top5_red['free sulfur dioxide'], width=0.2, color='blue', align='center', label='Free Sulfur Dioxide')

#Plot the shifted data as red bars; shifted right by 0.2 and increased by 50 in height
plt.bar(np.array(wine_nums) + 0.25, np.array(top5_red['total sulfur dioxide']), width=0.2, color='red', align='center', label='Sulfur Dioxide')

center_position = np.array(wine_nums) + 0.125
plt.xticks(center_position, [f'Wine {i}' for i in wine_nums])

#Add title and axis labels for better context
plt.title("Top 5 Red Wine Comparison of Free Sulfur Dioxide and Total Sulfur Dioxide",fontsize=14)
plt.xlabel("Wine Number", fontsize=14)
plt.ylabel("Sulfer Dioxide(ppm)",fontsize=14)

#Add grid and legend to the plot
plt.grid(True)
plt.legend()

#Optionally, annotate the bars to explain further details
for x, h in zip(wine_nums, top5_red['free sulfur dioxide']):
    plt.text(x, h, f"{h}", ha='center', va='bottom', color='blue')

for x, h in zip(np.array(wine_nums) + 0.25, np.array(top5_red['total sulfur dioxide'])):
    plt.text(x, h, f"{h}", ha='center', va='bottom', color='red')

#Display the plot
plt.tight_layout() #Fix plot layout
#Save the figure
fig.savefig('VicenteGarcia_Top5RedWineComparisonofFreeSulfurDioxideandTotalSulfurDioxideBarChart.png')
plt.show()


#Now for White Whine
#Assign the size of the plot figure
fig = plt.figure(figsize=(10, 6))
#Get only the values for the sulfer dioxide
plt.bar(wine_nums, top5_white['free sulfur dioxide'], width=0.2, color='blue', align='center', label='Free Sulfur Dioxide')

#Plot the shifted data as red bars; shifted right by 0.25
plt.bar(np.array(wine_nums) + 0.25, np.array(top5_white['total sulfur dioxide']), width=0.2, color='red', align='center', label='Sulfur Dioxide')

#Center the xticks to be inbetween the bars and add wine infront of the name for each top wine
center_position = np.array(wine_nums) + 0.125
plt.xticks(center_position, [f'Wine {i}' for i in wine_nums])

#Add title and axis labels for better context
plt.title("Top 5 White Wine Comparison of Free Sulfur Dioxide and Total Sulfur Dioxide",fontsize=14)
plt.xlabel("Wine Number", fontsize=14)
plt.ylabel("Sulfer Dioxide(ppm)",fontsize=14)

#Add grid and legend to the plot
plt.grid(True)
plt.legend()

#Optionally, annotate the bars to explain further details
for x, h in zip(wine_nums, top5_white['free sulfur dioxide']):
    plt.text(x, h, f"{h}", ha='center', va='bottom', color='blue')

for x, h in zip(np.array(wine_nums) + 0.25, np.array(top5_white['total sulfur dioxide'])):
    plt.text(x, h, f"{h}", ha='center', va='bottom', color='red')

# Display the plot
plt.tight_layout() #Fix plot layout
#Save the figure
fig.savefig('VicenteGarcia_Top5WhiteWineComparisonofFreeSulfurDioxideandTotalSulfurDioxideBarChart.png')
plt.show()
```

### Observations

The two graphs displayed show the different amounts of free sulfur dioxide and total sulfur dioxide between the top five wines in terms of quality for red and white wines. What these bar graphs seem to show is that in terms of free sulfur dioxide content, it seems that the top five red wines have less than the top five white wines, with most of them being around 10.0 ppm for red wine and 20.0 ppm for white wine. It also seems that white wines have a higher amount of total sulfur dioxide in the top five white wines, with the highest wine containing 139.0 ppm, and for red wine it is 88.0 ppm. These observations show us that for high-quality white wines in terms of sulfur dioxide, they have a higher amount of sulfur dioxide than red wines.


### Line Graph

```python
#Group the red and white wines by quality and get the average values for all the other attributes
sorted_redwine = redwine_tips.groupby('quality').mean().reset_index() #reset_index used to make pH a regular colum
sorted_whitewine = whitewine_tips.groupby('quality').mean().reset_index()

#Figure size
fig = plt.figure(figsize=(10,6))
#plot figure
plt.plot(sorted_redwine['quality'], sorted_redwine['alcohol'], color='red', marker='x', linestyle='dashed', label='Red Wine')
plt.plot(sorted_whitewine['quality'], sorted_whitewine['alcohol'], color='blue', marker='o', linestyle='solid', label='White Wine')

#Plot title
plt.title('Analysis Between Average Alcohol Content and Wine Quality', fontsize=14)
#Set y label
plt.ylabel('Alcohol Content', fontsize=13)
#Set x label
plt.xlabel('Wine Quality', fontsize=13)

plt.legend()
plt.grid(True)
plt.show()

#Save the figure
fig.savefig('VicenteGarcia_AnalyzetheAmountofAlcoholContentBasedonWineQualityLineGraph.png')
```

### Observation

This line graph is done to show the relationship between the quality of a wine and its average alcohol content for each quality of all the red and white wines in the data set. This graph seems to show that red wines seem to have a little bit more alcohol content than white wine throughout all the qualities except for 3. This graph also seems to show a linear relationship between the quality of the wine and the amount of alcohol content. These observations show us that when it comes to a wine's rating in terms of its physicochemical makeup, the higher the alcohol content, the higher its quality score will be.

```python
#Create the arrays for the pie chart
#Count the amount for each wine quality
redwine_unfilt = redwine_tips['quality'].value_counts().sort_index()
whitewine_unfilt = whitewine_tips['quality'].value_counts().sort_index()
#Filter the data so that only the wine qualities over 4 are shown
redwine_Quality = redwine_unfilt[redwine_unfilt.index > 4]
whitewine_Quality = whitewine_unfilt[whitewine_unfilt.index > 4]

#Make the labels for the different wine qualities
labels = redwine_Quality.index  # quality levels

fig = plt.figure(figsize=(7,7))
#Add the title
plt.title('Quality Amount in Red Wine', fontsize=16)
plt.pie(redwine_Quality, labels=labels, autopct='%1.1f%%');

plt.show()
#Save the figure
fig.savefig('VicenteGarcia_QualityAmountinRedWinePieChart.png')

#Make the labels for the different wine qualities
labels = whitewine_Quality.index  # quality levels

fig = plt.figure(figsize=(7,7))
#Add the title
plt.title('Quality Amount in White Wine', fontsize=16)
plt.pie(whitewine_Quality, labels=labels, autopct='%1.1f%%');
plt.show()

#Save the figure
fig.savefig('VicenteGarcia_QualityAmountinWhitewWinePieChart.png')
```

### Observation

These pie charts are designed to show the amount of quality wines above a quality of 5 that red wines and white wines contain in the data. This is done to show what color of wine seems to score better in terms of quality due to its  physicochemical makeup. In the pie chart for red wines, it seems to not contain a wine with the quality score of nine, as opposed to the white wine, which does. Also, red wines in the data set seem to be mostly made up of wines with a quality score of five, with it taking up 44.3% of the data, and in white wine, a quality score of 6 seems to be in the majority, with it making up 46.6% of the data. In general it seems that white wines also seem to have a higher count for each of the other quality scores except for five. These observations show us that white wines seem to score better in terms of quality in general than red wines.

### Box Plot

```python
#labels for each Wine Quality
wine_colors = ['Red Wine', 'White Wine']

#Prepare Data for the diffeerent levels of ph for the data of red and white wine
data_to_plot = [redwine_tips['pH'], whitewine_tips['pH']]

#Create a figure and plot the box plot
fig = plt.figure(figsize=(8, 6))
plt.boxplot(data_to_plot, tick_labels=wine_colors)
plt.title("Boxplot of pH Level by Wine Colors")
plt.xlabel("Wine Color")
plt.ylabel("pH")
plt.grid()
plt.show()

#Save the figure
fig.savefig('VicenteGarcia_BoxplotofpHLevelbyWineColorsBoxPlot.png')
```

### Observations

The Box plot above is done to show the difference in pH levels when it comes to all red and white wines  in the dataset this is done to show the way that the type of wine can change the amount of pH inside it. The box plot for red wine seems to show a higher amount of pH than white wines with its median being around a pH level of 3.3 and for white wines a little lower than 3.2 which is not much of a difference but somewhat significant. Also when it comes to red wine looking at the outliers in the data they seem to also reach a high pH level of 4.0 and white wine only reaching a little higher than 4.0. White wines also seem to be a little more evenly distributed due to it have a high amount of outliers in the 3.5 to 3.6 range. These observations seem to be pointing to white wines having a lower pH and more evenly disprused pH level as opposed to red wines.
